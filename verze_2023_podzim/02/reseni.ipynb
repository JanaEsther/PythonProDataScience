```python
import pandas as pd

data = pd.read_csv("traffic.csv")
data.head()
```

```python
data = data[data["Junction"] == 1]
data.head()

data["DateTime"] = pd.to_datetime(data["DateTime"])
data["DayOfWeek"] = data["DateTime"].dt.weekday
data["Hour"] = data["DateTime"].dt.hour
data.head()
```

```python
import numpy

data_pivot = pd.pivot_table(data, values="Vehicles", index="Hour", columns="DayOfWeek", aggfunc=numpy.mean, margins=True)
data_pivot
```

```python
bins = [0, 5, 10, 14, 17, 22, 24]
labels = ["noc_1", "ranní špička", "poledne", "odpolední špička", "večer", "noc_2"]
data['HourGroup'] = pd.cut(data["Hour"], bins=bins, labels=labels, right=False)
data["HourGroup"] = data["HourGroup"].str.replace("_1", "")
data["HourGroup"] = data["HourGroup"].str.replace("_2", "")
bins = [1, 5, 7]
labels = ["weekday", "weekend"]
data['DayOfWeekGroup'] = pd.cut(data["DayOfWeek"], bins=bins, labels=labels, right=False)
data.head(15)
```

```python
data_pivot = pd.pivot_table(data, values="Vehicles", index="HourGroup", columns="DayOfWeekGroup", aggfunc=numpy.mean, margins=True)
data_pivot
```
