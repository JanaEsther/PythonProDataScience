```python
import pandas as pd
df = pd.read_csv("half_marathon.csv")
df.head()
```

```python
df = df.sort_values(["Jmeno", "Rocnik", "Rok zavodu"])
df["Cas"] = pd.to_datetime(df["Cas"])
df["Cas 2020"] = df.groupby(["Jmeno", "Rocnik"])["Cas"].shift(-1)
df.to_csv("half_marathon_hint_1.csv", index=False)
df = df.dropna()
df["Zmena"] = df["Cas 2020"] - df["Cas"]
df.to_csv("half_marathon_hint_2.csv", index=False)
df["Zmena"] = df["Zmena"].dt.total_seconds()
df.to_csv("half_marathon_hint_3.csv", index=False)
df["Zmena"].mean()
```

```python
# Bonus
df["Zmena Text"] = pd.cut(df["Zmena"], bins=[float("-inf"), 0 , float("inf")], labels=["zlepsil se", "zhorsil se"])
df
```
