### Otázka 8

Nyní rozdělme jednotlivé studenty a studentky do skupin v závislosti na výsledku  (sloupec TOTAL SCORE (OUT of 100)). Napiš příkaz na vytvoření sloupce student group, kde budou následující hodnoty:

Need Improvement pro výsledek 0 až 25
Below Average pro výsledek 26 až 40
Average pro výsledek od 41 do 60
Above Average pro výsledek od 61 do 89
Top Achievers pro výsledek od 90 do 100 bodů

````python
score = [0, 25, 40, 60, 89, float("inf")] 
labels = ["Need Improvement", "Below Average", "Average", "Above Average", "Top Achievers"]
data["student_group"] = pd.cut(data["TOTAL SCORE (OUT of 100)"], bins=score, labels=labels)
data.head()
````

### Otázka 9

Nyní porovnejme výsledky z obecné zkoušky a zkoušky zaměřené na danou specializaci. Vytvoř sloupec test comparsion, který bude obsahovat hodnotu "better general" pro studenty, kteří měli více bodů z obecné zkoušky (sloupec  
GENERAL MANAGEMENT SCORE (OUT of 50) má vyšší hodnotu než 
Domain Specific SCORE (OUT 50)) a "better domain" pro ostatní. Tento postup je trochu matoucí pro studenty a studentky, kteří mají stejně bodů z obou. Pokud ti to vadí, můžeš zvolit těžší variantu a rozdělit je do skupin "better general", "equal", a "better domain".

*Pokud jste rozdělovali pouze na "better general" a "better domain" (a ti, kteří mají stejně, padli do jedné z těchto dvou skupin), strhl jsem vám čtvrt bodu.*

````python
data["test comparsion"] = np.where(data["GENERAL MANAGEMENT SCORE (OUT of 50)"] > data["Domain Specific SCORE (OUT 50)"], "better general", 
                                   np.where(data["GENERAL MANAGEMENT SCORE (OUT of 50)"] < data["Domain Specific SCORE (OUT 50)"], "better domain", "equal"))
````
