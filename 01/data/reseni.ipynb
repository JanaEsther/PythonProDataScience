```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("../hotel_bookings.csv")
```

```python
data_stays_week = data[data["stays_in_week_nights"] < 10]
```

```python
ax = sns.countplot(data=data_stays_week, x="stays_in_week_nights", hue="is_canceled")
ax.set(xlabel="Stays in week nights", ylabel="Count of reservations", title="Stays in week nights")
ax.legend(labels=["no", "yes"], title="Canceled")
```

```python
data["total_length"] = data["stays_in_weekend_nights"] + data["stays_in_week_nights"]
data_total_length = data[(data["total_length"] < 10) & (data["total_length"] > 0)]
```

```python
ax = sns.countplot(data=data_total_length, x="total_length", hue="is_canceled")
ax.set(xlabel="Stays in week nights", ylabel="Count of reservations", title="Stays in week nights")
ax.legend(labels=["no", "yes"], title="Canceled")
```

```python
data["country"].value_counts()
```

```python
data_adr = data[(data["adr"] < 300) & (data["country"] == "GBR")]
sns.histplot(data=data_adr, x="adr", hue="is_canceled", multiple="stack")
```
