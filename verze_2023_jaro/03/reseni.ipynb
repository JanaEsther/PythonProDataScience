```python
import pandas
df = pandas.read_csv("half_marathon.csv")

df = df.sort_values(["Jmeno", "Rok zavodu"])
df["Cas zavodnika"] = pandas.to_datetime(df["Cas zavodnika"])
df["Cas zavodnika 2020"] = df.groupby("Jmeno")["Cas zavodnika"].shift(-1)
df.to_csv("half_marathon_hint_1.csv", index=False)
df = df.dropna()
df["Zmena"] = df["Cas zavodnika 2020"] - df["Cas zavodnika"]
df.to_csv("half_marathon_hint_2.csv", index=False)
df["Zmena"] = df["Zmena"].dt.total_seconds()
df.to_csv("half_marathon_hint_3.csv", index=False)
df["Zmena"].mean()
```

```python
# Bonus
df["Zmena Text"] = pandas.cut(df["Zmena"], bins=[float("-inf"), 0 , float("inf")], labels=["zlepsil se", "zhorsil se"])
df
```
